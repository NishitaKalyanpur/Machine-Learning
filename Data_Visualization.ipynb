# Review: Linear Algebra and Probability Theory

### A. Linear Algebra

Linear Algebra is a branch of mathematics that studies linear algebric objects concerning linear equations which includes linear functions  (linear regression is an example: y = Ax). Their representation or rather the primitive objective is to study vector spaces and linear maps between the spaces i.e. they are represented through matrices and vector spaces. Linear algebra includes mathematical objects such as vectors, scalars, matrices, tensors and along with that multiplication, transformation, transpose, inverse of matrices etc.

I. Scalar, Vectors, Matrices, Tensors:

Scalar: It is just a single number. Usually written in italics and lowercase.

Vectors: It is an array of numbers. Each number is arranged in an order and can be referred from its index. Typically these vectors are written in lowercase.

Matrices: They are a 2D array or arrays with 2 or more dimensions. So each dimension has its own index. They are typically written in uppercase with bold typeface.

Tensors: An array of numbers arranged on a regular grid with avariable number of axes is known as a tensor.

II. Operations on Matrices:

1. Transpose: The transpose of a matrix is a new matrix whose rows are the columns of the original. (This makes the columns of the new matrix the rows of the original). 

2. Multiplication (Also applies to vectors): The matrix product of matrices A and B results in a third matrix C. Inorder for this product to be deﬁned, A must have the same number of columns as B has rows. If A is of shape m × n and B is of shape n × o, then C is of shape m × o. 

3. Inverse matrix: To describe matrix inversion, it is necessary to know what identity matrix is. It is a matrix that doesn't change any vector no even after multiplying that vector to other matrix. All the entries along with the diagonal values are set to 1. The product of inverse matrix with that matrix is always going to to the identity matrix.

Furthermore, there are different types of matrices such as a diagonal matrix, symmetric matrix, orthogonal matrix, etc.

Symmetric matrix: A symmetric matrix is any matrix that is equal to its own transpose.

Diagonal matrix: Diagonal matrices consist mostly of zeros and have nonzero entries only along the main diagonal.

Norm: In machine learning, the measurement of the size of vectors using a function called a norm. On an intuitive level, the norm of a vector x that measures the distance from the origin to the point x.

Eigendecomposition: It is a way of decomposing a matrix into a set of eigenvectors and eigenvalues. 

Determinant: The determinant of a square matrix, denoted det(A), is a value that can be computed from the elements of the matrix. 




### B. Probability Theory

Probability theory is a mathematical framework for representing uncertainstatements. It plays an important role in machine learning, , as the design of learning algorithms often relies on probabilistic assumption of the data. It provides a means of quantifying uncertainty as well as axioms forderiving new uncertain statements.

1. Probability

When we speak about probability, we often refer to the probability of an event of uncertain nature taking place. For example, we speak about the probability of rain next Tuesday. Many branches of computer science deal mostly with entities that are entirely deterministic and certain. Probability can be seen as the extension of logic to deal with uncertainty. Logic provides a set of formal rules for determining what propositions are implied to be true or false given the assumption that some other set of propositions is true or false. Probability theory provides a set of formal rules for determining the likelihood of a proposition being true given the likelihood of other propositions.

2. Random Variables

A random variable is a variable that can take on diﬀerent values randomly. Random variables may be discrete or continuous. A discrete random variableis one that has a ﬁnite or countably inﬁnite number of states. Note that these states are not necessarily the integers; they can also just be named states that are not considered to have any numerical value. A continuous random variable is associated with a real value. They are functions that map outcomes (in the outcome space) to real values.

3. Distributions, Joint Distributions, and Marginal Distributions

A probability distribution is a description of how likely a random variable or set of random variables is to take on each of its possible states. This formally refers to the probability of a random variable taking on certain values. Sometimes, we speak about the distribution of more than one variables at a time. We call these distributions joint distributions, as the probability is determined jointly by all the variables involved. The name of marginal distribution comes from the fact that if we add up all the entries of a row (or a column) of a joint distribution, and write the answer at the end (i.e., margin)
of it, this will be the probability of the random variable taking on that value. 

4. Discrete Distributionn: Probability Mass Function

By a discrete distribution, we mean that the random variable of the underlying distribution can take on only finitely many different values (or that the outcome space is finite).

5.  Continuous Distribution: Probability Density Function

By a continuous distribution, we mean that the random variable of the underlying distribution can take on infinitely many different values (or that the outcome space is infinite).

6. Chain Rule and Bayes Rule

The Chain Rule is often used to evaluate the joint probability of some random variables, and is especially useful when there are (conditional) independence across variables. 

The Bayes Rule allows us to compute the conditional probability P(X|Y ) from P(Y |X), in a sense “inverting” the conditions. 

P (x | y) = ( P (x) P (y | x) )/P (y)

This dataset was derived from geospatial data from two sources: 1) Landsat time-series satellite imagery from the years 2014-2015, and 2) crowdsourced georeferenced polygons with land cover labels obtained from OpenStreetMap. The crowdsourced polygons cover only a small part of the image area, and are used used to extract training data from the image for classifying the rest of the image.

###### Data Attribute Information:
class: The land cover class (impervious, farm, forest, grass, orchard, water). 

max_ndvi: the maximum NDVI (normalized difference vegetation index) value derived from the time-series of satellite images. 

20150720_N - 20140101_N : NDVI values extracted from satellite images acquired between January 2014 and July 2015, in reverse chronological order (dates given in the format yyyymmdd).
